```python
import pymongo
import time
import string
import re
import os
import json
from bson import json_util
```

```python
MongoClient = pymongo.MongoClient
conn = MongoClient('mongo-enron', 27017)
#use enron database
db = conn['enron']
messages = db.messages
messageswords = db.messageswords
topics = db.topics
authortopics = db.authortopics


def get_topics():
    rows = topics.find()
    return rows  

def get_author_topics(topicname):
    rows = authortopics.find({}).sort([(topicname, -1)]).limit(6)
    return rows

def get_message(id):
    rows = messages.find({"message-id":id},{"body":1, "filename":1, "from":1, "x-from":1, "date":1, "to":1, "subject":1})
    return rows;

def get_message_list(author):
    pipeline = [
        {"$match": {"from": author}},
        {"$lookup":{"from": "messages", "localField": "message-id", "foreignField": "message-id","as": "message"}},
        {"$unwind": "$message"},
        {"$project": {"body":"$message.body", "from": "$from", "message-id": "$message-id", "date": "$message.date", "subject": "$message.subject"}}
    ]
    rows = list(messageswords.aggregate(pipeline, allowDiskUse=True))
    return rows
    

#def main():
#     response = get_message('<15464986.1075855378456.JavaMail.evans@thyme>')
#     json_response = json.dumps(list(response),  default=json_util.default)
#     #print json_response

#     print "-------------"
#     response = get_topics()
#     json_response = json.dumps(list(response),  default=json_util.default)
#     #print json_response

#     print "--------------"
#     response = get_author_topics("topic2")
#     json_response = json.dumps(list(response),  default=json_util.default)
#     #print json_response

    
#     print "--------------"
#     response = get_message_list("sylvia.hu@enron.com")
#     json_response = json.dumps(list(response),  default=json_util.default)
#     print json_response

    
#     print 'hi'
    
#main()
```
